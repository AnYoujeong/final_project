# Remove Dot

```python
df_final = pd.read_csv('/content/sample_data/final_chord_LSTM.csv')
df_final2 = df_final.fillna('.')



# 데이터 전처리 

# 전체 코드에 들어있는 코드들 사전 정의

code_all = []
chord_all2 = []

for i in range(len(df_final2)) :
    df_new = df_final2.iloc[i].to_list()
    count={}

    for i in df_new :
        try: count[i] += 1
        except: count[i]=1
    new_list = []
    for v in df_new:
        if v not in new_list:
            new_list.append(v)

    # 코드 하나씩 분리 
    code = str(new_list).split(',')
    #print("0 : ",code)
    code = [x for x in code if str(x) != 'nan']

    # 공백제거
    code_1 = list(map(lambda s : ''.join(s.split()), code))
    # ' 제거
    code_2=str(code_1).replace("'","")

    code_3 = code_2.replace("[","")  # str에서 대괄호 없애기
    code_4 = code_3.replace("]","")  # ]없애기
    code_5 = code_4.replace('"',"") # " 없애기
    code_6 = code_5.split(",")  # ,로 나누기
    code_7 = list(map(lambda s : ''.join(s.split()), code_6)) # 공백 한번 더 제거

    chord_all2.append(code_7)
    for t in code_7:
        code_all.append(t)



# '.' 제거
chord_all3 = []

for l in chord_all2:
    if '.' in l:
        l.remove('.')
        print('remove dot')
    else:
        print('no dot')
    chord_all3.append(l)
print(chord_all3)
    
```

```python
#list to csv

import csv

with open('chord_all3.csv', 'w', newline='') as f:
    write = csv.writer(f)
    write.writerows(chord_all3)
```

```python
print('총 곡 수 :{}'.format(len(chord_all3)))
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(chord_all3)
vocab_size = len(tokenizer.word_index) + 1
print('단어 집합의 크기 : %d' % vocab_size)
```

```python
sequences = list()

for sentence in chord_all3:

    # 각 샘플에 대한 정수 인코딩
    encoded = tokenizer.texts_to_sequences([sentence])[0] 
    for i in range(3, len(encoded)):
        sequence = encoded[:i+1]
        sequences.append(sequence)

sequences[:5]
```

```python
max_len = max(len(l) for l in sequences)
print('샘플의 최대 길이 : {}'.format(max_len))
```

```python
# 가장 긴 샘플의 길이인 67로 모든 샘플의 길이를 패딩
# padding='pre'를 설정하여 샘플의 길이가 24보다 짧은 경우에 앞에 0으로 패딩
sequences = pad_sequences(sequences, maxlen=max_len, padding='pre')
print(sequences[:3])
```

```python
sequences = np.array(sequences)
X = sequences[:,:-1]
y = sequences[:,-1]
print(X[:3])
print(y[:3])
```

```python
y = to_categorical(y, num_classes=vocab_size)
```

```python
embedding_dim = 10
hidden_units = 128

model = Sequential()
model.add(Embedding(vocab_size, embedding_dim))
model.add(LSTM(hidden_units))
model.add(Dense(vocab_size, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=200, verbose=2)
```

```python
def sentence_generation(model, tokenizer, current_word, n): # 모델, 토크나이저, 현재 단어, 반복할 횟수
    init_word = current_word
    sentence = ''

    # n번 반복
    for _ in range(n):
        encoded = tokenizer.texts_to_sequences([current_word])[0]
        encoded = pad_sequences([encoded], maxlen=max_len-1, padding='pre')

        # 입력한 X(현재 단어)에 대해서 y를 예측하고 y(예측한 단어)를 result에 저장.
        result = model.predict(encoded, verbose=0)
        result = np.argmax(result, axis=1)

        for word, index in tokenizer.word_index.items(): 
            # 만약 예측한 단어와 인덱스와 동일한 단어가 있다면
            if index == result:
                break

        # 현재 단어 + ' ' + 예측 단어를 현재 단어로 변경
        current_word = current_word + ' '  + word

        # 예측 단어를 문장에 저장
        sentence = sentence + ' ' + word

    sentence = init_word + sentence
    return sentence
```
